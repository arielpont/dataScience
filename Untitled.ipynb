```python
# haciendo gr√°ficos de burbujas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

df = pd.read_csv("csv/flights.csv", nrows = 100000)
df = df.sample(frac=1).head(1000)
```

```python
plt.scatter(x = df.DAY_OF_WEEK, y = df.ARRIVAL_DELAY, s = df.DISTANCE)
```

```python
plt.scatter(x = df.DAY_OF_WEEK, y = df.ARRIVAL_DELAY, s = df.DISTANCE, alpha = 0.3)
plt.title("T√≠tulo de prueba")
plt.xlabel("X")
plt.ylabel("Y")
plt.ylim([0, 150])
# plt.text()
```
