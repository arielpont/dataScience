```python
import pandas as pd
df = pd.read_csv("csv/Sample10K.csv", nrows = 1000)
```

```python
dfduplicate = df.append(df)
```

```python
dfduplicate = dfduplicate.sample(frac=1)
```

```python
dfclean = dfduplicate.drop_duplicates()
```

```python
len(dfclean) == len(df)
```

```python
# elimina dilas que tienen valores duplicados según una columna
dfclean.drop_duplicates(subset = "Employee Markme")
```

```python
# gestionando las filas con datos faltantes
df.dropna()
```

```python
# va a filtrar filas que tengan como máximo 2 datos faltantes de los 5
df.dropna(thresh=2)
```

```python
# de una forma más dinámica
df.dropna(thresh=len(df.columns) - 2)
```

```python
df.dropna(subset = ["Employee Markme"])
```
