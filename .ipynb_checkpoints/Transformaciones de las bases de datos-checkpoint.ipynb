```python
# una "transformación" es afectar la estructura de la base de datos. Agregar columnas por ejemplo.

import pandas as pd
df = pd.read_csv("csv/Sample10K.csv", nrows = 1000)
```

```python
# creando una nueva columna
df["LeaveX2"] = round(df["Leave"] * 2)
```

```python
df["LeaveX2"].head()
```

```python
# eliminando una columna
del(df["LeaveX2"])
```

```python
df.head()
```

```python
# borrar más de una columna (recordar que debemos igual para que se guarde)
# df = df.drop()
# se puede usar al final "inplace" para reemplazar por un valor específico
df.drop(["Serial Number", "Company Name"], axis=1)
# df.drop(["Serial Number", "Company Name"], axis=1, inplace=True)
```

```python
# elimina la primer fila
df.drop(0)
```

```python
# eliminar un rango de filas
df.drop(range(0, 100))
```

```python
dfLeave = df[df.Leave == 0]
dfLeave1 = df[df.Leave == 1]
```

```python
# creando un nuevo dataframe a partir de otros que tienen la misma estructura para distintos datos
newdf = dfLeave.append(dfLeave1)
```
