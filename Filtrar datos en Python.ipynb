```python
import pandas as pd
df = pd.read_csv("csv/Sample10K.csv", nrows = 1000)
```

```python
# obtener los valores de una sola columna
df["Company Name"].head(10)
```

```python
# para obtener valores por filas podemos usar
# recordar que Python no incluye el último valor
df[7:15]
```

```python
# si queremos obtener valores específicos bajo condiciones
df[df["Leave"] != 0]
```

```python
# filtros combinados
df[
    (df["Employee Markme"] == "Mark") &
    (df["Leave"] != 0)
    ].head()
```

```python
# otra sintaxis para filtrar por más valores dentro de una columna
df[df.Leave.isin([1, 2])].head()
```

```python
# ver los primeros 5 casos donde no hay valores, osea NaN
df[pd.isna(df["Employee Markme"])].head()
```

```python
# conocer el total de filas que no tienen valor asignado en una columna específica
len(df[pd.isna(df["Employee Markme"])])
```
