```python
import pandas as pd
df = pd.read_csv("csv/Sample10K.csv", nrows = 1000)
```

```python
df.groupby(by = "Company Name")["Leave"].max()
```

```python
# muy interesante, tira estádística básica
df.groupby(by = "Company Name")["Leave"].describe()
```

```python
# calculando la media
df.groupby(by = "Company Name")["Leave"].mean()
```

```python
# podemos hacer operaciones y calcular datos
df.groupby(by = "Company Name")["Leave"].max() - df.groupby(by = "Company Name")["Leave"].min()
```

```python
# los GROUPBY se pueden guardar como objetos y luego aplicar distintas funciones max, min, describe, etc
dfgroupby = df.groupby(by = "Company Name")["Leave"]
```

```python
dfgroupby.max()
```

```python
dfgroupby.describe()
```
